```python
weatherDataCP = pd.read_csv("Data/tempDataCP.csv", sep = ",",encoding = "ISO-8859-1",  low_memory= True, skiprows=0, nrows=None)
weatherDataJFK = pd.read_csv("Data/tempDataJFK.csv", sep = ",",encoding = "ISO-8859-1",  low_memory= True, skiprows=0, nrows=None)
weatherDataLGA = pd.read_csv("Data/tempDataLGA.csv", sep = ",",encoding = "ISO-8859-1",  low_memory= True, skiprows=0, nrows=None)
```

```python
weatherDataCP.head()
```

```python
weatherDataJFK.head()
```

```python
weatherDataLGA.head()
```

```python
weatherDataCP.drop(['TimeEST', 'TimeEDT'], axis=1, inplace=True)
weatherDataJFK.drop(['TimeEST', 'TimeEDT'], axis=1, inplace=True)
weatherDataLGA.drop(['TimeEST', 'TimeEDT'], axis=1, inplace=True)


weatherDataCP = weatherDataCP.rename(columns={'Unnamed: 0': 's.no.'})
weatherDataJFK = weatherDataJFK.rename(columns={'Unnamed: 0': 's.no.'})
weatherDataLGA = weatherDataLGA.rename(columns={'Unnamed: 0': 's.no.'})
```

```python
frames = [weatherDataCP,weatherDataJFK,weatherDataLGA]
weatherData = pd.concat(frames)
weatherData.head()
weatherData.loc[(weatherData['Wind_SpeedMPH'] == 'Calm'), ['Wind_SpeedMPH']] = '0'
weatherData["Wind_SpeedMPH"] = weatherData["Wind_SpeedMPH"].astype(np.float64)
```

```python
weatherData = weatherData.groupby("Time", as_index=False).mean()
weatherData.head()
```
